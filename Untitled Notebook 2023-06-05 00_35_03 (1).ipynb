```
#To read all files

file1 = spark.read.csv("dbfs:/FileStore/tables/AppleStore1-1.csv", header=True, inferSchema=True)
file2 = spark.read.csv("dbfs:/FileStore/tables/AppleStore2-1.csv", header=True, inferSchema=True)
file3 = spark.read.csv("dbfs:/FileStore/tables/AppleStore3-1.csv", header=True, inferSchema=True)
file4 = spark.read.csv("dbfs:/FileStore/tables/AppleStore4-1.csv", header=True, inferSchema=True)
```

```
#Merge the datasets to create a unified dataset for analysis

merged_dataset = file1.unionAll(file2).unionAll(file3).unionAll(file4)
merged_dataset.show()
```

```
#To find duplicates values

from pyspark.sql.functions import col

duplicates = merged_dataset.groupBy(merged_dataset.columns).count().filter(col("count") > 1)
duplicates.show()
```

```
#To find the null values

from pyspark.sql import SparkSession
from pyspark.sql.functions import col, isnan, sum


spark = SparkSession.builder.getOrCreate()


null_counts = merged_dataset.select([sum(col(c).isNull().cast("int")).alias(c) for c in merged_dataset.columns])

null_counts.show()
```
